```python
import pandas as pd
import numpy as np
from filter_rules_apriori import read_file
from filter_rules_apriori import filter as filter_rules
```

```python
df_dataset_apriori = pd.read_csv('storage/dataset_apriori.csv')

df_count = df_dataset_apriori['PERMANENCIA_PROLONGADA'].value_counts()

print(df_count)

print('False', df_count[False]/df_dataset_apriori.shape[0])
print('True', df_count[True]/df_dataset_apriori.shape[0])
```

```python
rules = read_file('storage/rules_apriori.txt')
df_rules = filter_rules(rules)
df_rules
```

```python
df_rules['coverage'] = df_rules['coverage'].astype(float)
df_rules['confidence'] = df_rules['confidence'].astype(float)
df_rules['lift'] = df_rules['lift'].astype(float)

df_rules = df_rules.sort_values(['lift', 'confidence', 'coverage'], ascending=False).reset_index(drop=True)

df_rules
```

```python
df_positive = df_rules[df_rules['implies'] == 'PERMANENCIA_PROLONGADA=True'].drop(['implies'], axis=1)
df_positive.sort_values(['lift'], ascending=False).head(50)
```

```python
df_negative = df_rules[df_rules['implies'] == 'PERMANENCIA_PROLONGADA=False'].drop(['implies'], axis=1)

df_negative[
    ~(df_negative['rule'].str.contains('BOLSA_ESTUDANTIL=True') &
    df_negative['rule'].str.contains('TRABALHO_DURANTE_GRAD=False') |
    df_negative['rule'].str.contains('PRINCIPAL_MOTIVACAO=True')) &
    (df_negative['confidence'] > 0.73) &
    (df_negative['coverage'] > 0.15)
].sort_values(['lift'], ascending=False).head(50)

# df_negative[40:].head(50)
```
